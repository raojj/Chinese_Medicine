```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7,8]
y = [60,45,49,36,42,67,40,50]
plt.bar(x,y)
```

```python
# 调整柱形图中每根柱子的宽度和颜色
plt.bar(x,y,width=0.5,color='k')
```

```python
# 绘制饼图
x = ['上海','成都','重庆','深圳','北京','长沙','南京','青岛']
y = [60,45,49,36,42,67,40,50]
plt.pie(y,labels = x,labeldistance = 1.1, autopct = '%.2f%%', pctdistance = 1.5)
```

```python
plt.pie(y,labels = x,labeldistance = 1.1, autopct = '%.2f%%', pctdistance = 1.5,explode = [0,0,0,0,0,0.1,0,0],startangle = 90, counterclock = False)
```

```python
# 绘制环形图
plt.pie(y,labels = x,labeldistance = 1.1, autopct = '%.2f%%', pctdistance = 1.5,wedgeprops = {'width':0.3,'linewidth':2,'edgecolor':'white'})
```
