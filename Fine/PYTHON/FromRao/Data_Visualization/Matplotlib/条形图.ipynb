```python
import matplotlib
import numpy as np
matplotlib.use('nbagg')
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(0)
x = np.arange(5)
y = np.random.randint(-5,5,5)
fig,axes = plt.subplots(ncols=2)
v_bars = axes[0].bar(x,y,color ='black')
h_bars = axes[1].barh(x,y,color ='black')

axes[0].axhline(0,color='r',linewidth=2.0)
axes[1].axvline(0,color='b',linewidth=2.0)
```

```python
fig,ax = plt.subplots()
v_bars = ax.bar(x,y,color='lightblue')
for bar,height in zip(v_bars,y):
    if height < 0:
        bar.set(edgecolor='darkred',color ='green',linewidth =3.0)
```

```python
x = np.random.randn(100).cumsum()
y = np.linspace(0,10,100)

fig,ax = plt.subplots()
ax.fill_between(x,y,color='lightblue')
```

```python
x = np.linspace(0,10,200)
y1 = 2*x + 1
y2 = 3*x + 1.2
y_mean = 0.5*x*np.cos(2*x)+2.5*x + 1.1
fig,ax = plt.subplots()
ax.fill_between(x,y1,y2,color='red')
ax.plot(x,y_mean,color='black')
```
